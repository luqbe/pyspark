```
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .appName("practice") \
    .getOrCreate()  
from pyspark.sql import SparkSession   
from pyspark.sql.types import StructType,IntegerType,StructField,StringType,DateType
schema=StructType([StructField("employee_id",IntegerType(),True),
                   StructField("employee_name",StringType(),True),
                   StructField("department",StringType(),True),
                   StructField("State",StringType(),True),
                   StructField("salary",IntegerType(),True), 
                   StructField("Age",IntegerType(),True)
                   ])

data= (
    (11,"james", "D101","ny",9000,34),
     (12,"michel","D101","ny",8900,32),
    (13,"robert", "D102","ca",7900,29),
    (14,"scott","D103","ca",8000,36),
    (15,"jen", "D102","ny",9500,38),
     (16,"jeff","D103","uk",9100,35),
    (17,"maria", "D101","ny",7900,40)
    )
df1=spark.createDataFrame(data,schema=schema)    
                   
```

```
from pyspark.sql import SparkSession   
from pyspark.sql.types import StructType,IntegerType,StructField,StringType,DateType
schema=StructType([StructField("dept_id",StringType(),True),
                   StructField("dept_name",StringType(),True)
])
data=(
    ["D101","sales"],
    ["D102","finance"],
    ["D103","marketing"],
    ["D104","Hr"],
    ["D105","support"]
)
df2=spark.createDataFrame(data,schema=schema)
                               
```

```
from pyspark.sql import SparkSession   
from pyspark.sql.types import StructType,IntegerType,StructField,StringType,DateType
schema=StructType([StructField("country_code",StringType(),True),
                   StructField("country_name",StringType(),True)
])
data=(
    ["ny","new_york"],
    ["ca","california"],
    ["uk","Russia"]

)
df3=spark.createDataFrame(data,schema=schema)
```

```
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,count,avg
df_avg=df1.groupBy("department").agg(avg(col("salary"))).alias("avg_salary")
df_avg.show()
```

```
from pyspark.sql.functions import col
df_employee_name=df1.filter(col("employee_name").like("m%")).alias("m_employees")\
                    .select(col("department"))
df_employee_name.show()                                     
                                       
```

```
from pyspark.sql.functions import col
df_bonus=df1.withColumn("bonus_salary",df1["salary"]*2)
df_bonus.show()
```

```
from pyspark.sql.functions import col
new_column_order = ["employee_id", "employee_name", "salary", "State", "Age", "department"]
df1=df1.select(new_column_order)
df1.show()
```

```
from pyspark.sql.functions import col
df_inner=df1.join(df2,df1["department"]==df2["dept_id"],"inner")\
            .select(col("employee_name"),col("department"),col("salary"))
df_inner.show()
df_left=df1.join(df2,df1["department"]==df2["dept_id"],"left")
df_left.show()
df_right=df1.join(df2,df1["department"]==df2["dept_id"],"right")
df_right.show()    
```

```
from pyspark.sql.functions import col 
df_rename=df1.withColumnRenamed("State","Country_name")
df_rename.show()
```

```
from pyspark.sql.functions import col,lower
df_lower = df_rename.toDF(*[col_name.lower() for col_name in df_rename.columns])
df_lower.show()
from pyspark.sql.functions import current_date

df_with_date = df_lower.withColumn("load_date", current_date())
df_with_date.show()
```
